## Simple Kriging

```python
def simple_kriging_method(data, unknown_index):
    
    data_corrupted = data[data.index != unknown_index]
    V = Variogram(data_corrupted[['X', 'Y']], data_corrupted['PORO'], model='gaussian', normalize=False)
    
    V.estimator = 'matheron'
    V.model = 'spherical'

    xdata = V.bins
    ydata = V.experimental

    cof, cov =curve_fit(models.spherical, xdata, ydata)
    ar1 = np.expand_dims(np.array(data[['X', 'Y']].loc[unknown_index]), axis = 0)
    ar2 = np.array(data_corrupted[['X', 'Y']])
    distance_matrix = pdist((np.concatenate((ar1, ar2), axis=0)))
    model = lambda h: spherical(h, cof[0], cof[1], cof[2])
    dists = pdist(ar2)
    
    M = squareform(model(dists))
    M = np.concatenate((M, [[1] * len(data_corrupted)]), axis=0)
    M = np.concatenate((M, [[1]] * len(data_corrupted) +[[0]]), axis=1)
    
    variances = model(distance_matrix[:len(data_corrupted)])
    B = np.concatenate((variances, [1]))
    
    weights = solve(M, B)
    obtained_perm = data_corrupted['PORO'].dot(weights[:-1])
    result.append(obtained_perm)
    return result
```

```python
result = []
for i in df.index:
    res= simple_kriging_method(df, i)
```

```python
df['kriging'] = res
```
